```python
#Разбивка по суммам грантов (сколько грантов относятся к каждой категории по сумме)
sum_grants = df['grantAmount'].to_list()
less1 =[]
between1_2 =[]
between2_3 = []
between3_4 = []
between4_5 = []
between5_6 = []
between6_7 = []
between7_8 = []
between8_9 = []
between9_10 = []
more10 = []
for i in sum_grants:
    i = int(i)
    if i<=1000000:
        less1.append(i)
    elif 1000000<i<=2000000:
        between1_2.append(i)
    elif 2000000<i<=3000000:
        between2_3.append(i)
    elif 3000000<i<=4000000:
        between3_4.append(i)
    elif 4000000<i<=5000000:
        between4_5.append(i)
    elif 5000000<i<=6000000:
        between5_6.append(i)
    elif 6000000<i<=7000000:
        between6_7.append(i)
    elif 7000000<i<=8000000:
        between7_8.append(i)
    elif 8000000<i<=9000000:
        between8_9.append(i)
    elif 9000000<i<=10000000:
        between9_10.append(i)
    elif 10000000<i:
        more10.append(i)

category = [less1,between1_2,between2_3,between3_4,between4_5,between5_6,between6_7,between7_8,between8_9,between9_10,more10]
for cat in category:
    print(len(cat))
```

```python
#Разбивка по суммам грантов для каждого оператора
operators = df['operatorName'].unique().tolist()
all_ops = []
for every_operator in operators:
    
    less1 =[]
    between1_2 =[]
    between2_3 = []
    between3_4 = []
    between4_5 = []
    between5_6 = []
    between6_7 = []
    between7_8 = []
    between8_9 = []
    between9_10 = []
    more10 = []
    for i in df[df['operatorName'].isin([every_operator])]['grantAmount'].to_list():
        i = int(i)
        if i<=1000000:
            less1.append(i)
        elif 1000000<i<=2000000:
            between1_2.append(i)
        elif 2000000<i<=3000000:
            between2_3.append(i)
        elif 3000000<i<=4000000:
            between3_4.append(i)
        elif 4000000<i<=5000000:
            between4_5.append(i)
        elif 5000000<i<=6000000:
            between5_6.append(i)
        elif 6000000<i<=7000000:
            between6_7.append(i)
        elif 7000000<i<=8000000:
            between7_8.append(i)
        elif 8000000<i<=9000000:
            between8_9.append(i)
        elif 9000000<i<=10000000:
            between9_10.append(i)
        elif 10000000<i:
            more10.append(i)

    category = [every_operator, len(less1),len(between1_2),len(between2_3),len(between3_4),len(between4_5),
                len(between5_6),len(between6_7),len(between7_8),len(between8_9),len(between9_10),len(more10)]
    all_ops.append(category)
res = pd.DataFrame(all_ops, columns = ['operator', 'less1','between1_2', 'between2_3',
                                      'between3_4','between4_5','between5_6','between6_7',
                                      'between7_8','between8_9','between9_10','more10'])
res
```
