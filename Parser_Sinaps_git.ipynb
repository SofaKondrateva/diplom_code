```python
from selenium import webdriver
from bs4 import BeautifulSoup
import requests
import time
import pandas as pd
```

```python
df = pd.read_csv('merge_2013-2019.csv', sep = ',', encoding = 'utf-8')
ogrn_list = df['receiverOGRN'].astype(str).unique().tolist()
searchurl = 'https://synapsenet.ru/searchorganization/proverka-kontragentov'
edges_set = []
```

```python
browser = webdriver.Chrome(executable_path = 'C:\\Users\\Sony\\diplom_parser\\chromedriver.exe')
browser.set_window_size(1280,1024) # задаем размер окна
```

```python
#Функция принимает на ОГРН НКО, а возвращает список всех связанных с организацией компаний
def get_all_connection_ogrns(ogrn):
    browser.get(searchurl)

    form = browser.find_element_by_id('org-search-input')
    form.send_keys(ogrn)
    button = browser.find_element_by_id('org-search-button') #Заполняем форму ОГРН’ом и отправляем
    button.click()
#Нажимаем на кнопки «показать всех», чтобы развернуть списки связанных компаний
    try:
        show_full_button_list = browser.find_elements_by_xpath("//div[@id='org-full-group-relationship']/div[@class='org-full-button']")
        for every_button in show_full_button_list:
            every_button.click()
            time.sleep(3)
    except:
        pass
    
    links = []
    conn_ogrn = []
    conn = browser.find_element_by_id('org-full-group-relationship')

    for link in conn.find_elements_by_class_name('org-link'): #Собираем ссылки на связанные компании
        link = link.get_attribute("href") 
        links.append(link)

    for every_link in links:
        browser.get(every_link)
        ogrn_parsed = browser.find_element_by_xpath("//div[@class='of-common-left']/div[@class='of-common-data'][2]/ul[@class ='ofcd-requisites']/li[1]").text
        ogrn_parsed = ogrn_parsed.strip('ОГРН — ')
        conn_ogrn.append(ogrn_parsed)
        time.sleep(3)
        
    return conn_ogrn
```

```python
#Функция принимает на вход список связанных организаций, проверяет, 
#есть ли такие ОГРН в датасете и возвращает список ОГРН, которые нашлись.
def check_grants(conn_ogrn):
    list_conn_ogrns = list(set(df[df['receiverOGRN'].isin(conn_ogrn)]['receiverOGRN']))
    return list_conn_ogrns
```

```python
#Функция строит ребра для графа. Принимает на вход ОГРН исходной компании и список связанных компаний, и записывает их в список.
def make_edges(ogrn,list_conn_ogrns):
    for every_conn in list_conn_ogrns:
        edges_set.append([ogrn,str(every_conn)])
```

```python
#Запускаем парсер
for ogrn in ogrn_list: 
    try:
        conn_ogrn = get_all_connection_ogrns(ogrn)
        make_edges(ogrn, check_grants(conn_ogrn))
    except:
        print ('Something wrong', ogrn)
        continue
```

```python
#Формируем датасет и записываем в csv-файл
df_edges = pd.DataFrame(edges_set, columns = ['Source', 'Target'])
```

```python
df_edges.to_csv('edges_full.csv', encoding='utf-8')
```
